```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 웹 드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 대상 웹 페이지 열기
driver.get('https://threads.net')  # 실제 사이트 URL로 변경

try:
    # "로그인" 버튼을 찾고 클릭
    login_button = driver.find_element(By.XPATH, '//a[@href="/login/?show_choice_screen=false"]')
    login_button.click()

    # 페이지 로딩을 기다리기 위한 시간 (필요에 따라 수정)
    time.sleep(3)

    # 사용자 이름 입력 필드 찾기 및 값 입력
    username_input = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//input[@autocomplete="username"]'))
    )
    username_input.clear()  # 기존 값을 지우고
    username_input.send_keys('o_ujung_')  # 사용자 이름 입력

    # 비밀번호 입력 필드 찾기 및 값 입력
    password_input = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//input[@autocomplete="current-password"]'))
    )
    password_input.clear()  # 기존 값을 지우고
    password_input.send_keys('dmdsk2dbwjd!')  # 비밀번호 입력

    # 첫 번째 로그인 제출 버튼 클릭
    login_submit_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//div[contains(@class,"x1i10hfl") and contains(@role,"button") and contains(., "로그인")]'))
    )
    login_submit_button.click()  # 첫 번째 버튼 클릭

    # 페이지 로딩을 기다리기 위한 시간 (필요에 따라 수정)
    time.sleep(5)

    # 로그인 후 페이지 확인
    print(f'현재 페이지 URL: {driver.current_url}')

except Exception as e:
    print(f'오류 발생: {e}')

try:
    # "검색" 링크 찾기 (href="/search" 속성으로 찾기)
    search_link = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//a[@href="/search"]'))
    )
    search_link.click()  # 검색 페이지로 이동

    # 페이지 로딩을 기다리기 위한 시간 (필요에 따라 수정)
    time.sleep(3)

    # 검색 페이지 확인
    print(f'현재 페이지 URL: {driver.current_url}')

except Exception as e:
    print(f'오류 발생: {e}')

# finally:
#     # 웹 드라이버 종료
#     driver.quit()
```
